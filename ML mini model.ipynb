```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = {'Hours': [1, 2, 3, 4, 5], 'Scores': [52, 98, 145, 210, 240 ]}
df = pd.DataFrame(data)

model = LinearRegression()
model.fit(df[['Hours']], df['Scores'])

pred = model.predict(pd.DataFrame({'Hours': [6]}))
print(pred)
```

```python
import matplotlib.pyplot as plt

# Plot actual data
# df, model, and pred are already defined in the previous cell
plt.scatter(df['Hours'], df['Scores'], label='Actual data')

# Plot regression line
plt.plot(df['Hours'], model.predict(df[['Hours']]), label='Best fit line')

# Plot prediction point
plt.scatter(6, pred[0], s=100, label='Prediction (6 hrs)')

# Labels
plt.title('Study Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.legend()

plt.show()
```
