```python
def simple_chatbot():
    print("Hello! I am a simple chatbot." 
          "How can i help you today?")
    while True: 
        user_input = input("You: ").lower()

        if "hello" in user_input: 
            print("Chatbot: Hi there!"
                  " How can i assist you today?")
        elif "how are you" in user_input: 
            print("Chatbot: I'm just a program,"
                  " but I'm doing fine!"
                  " How about you? ")
        elif "bye" in user_input:
            print(" Chatbot: Peace out!"
                  " Have a great day!")
            break
        elif "your name" in user_input:
            print("Chatbot: I'm a simple Chatbot, "
                  " Designed to help you. "
                  " You can call me Labubu!")
        elif "help" in user_input:
            print(" Chatbot: Sure!"
                  " You can ask me simple questions like 'Hello',"
                  " 'How are you?', or"
                  " 'What is your name?'") 
        else:
            print("Chatbot: I'm sorry, "
                  " I dont understand that."
                  " Can you try asking something else?")    
simple_chatbot()            
```
