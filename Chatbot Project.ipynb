```python
def course_assistant_bot():
    
    print("Hello! I am your Course Assistant Bot. ")
    print("Ask me anything about the AI course or university policies.\n")

    while True:
        user_input = input("You: ").lower()

        if "hello" in user_input:
            print("ChatBot: Hello! How can I help you today?")

        elif "course duration" in user_input:
            print("ChatBot: The AI course runs for 12 weeks, including lectures and projects.")

        elif "assignment deadline" in user_input:
            print("ChatBot: Assignments deadlines depend on the date or time set by the Lecturer.")

        elif "grading" in user_input:
            print("ChatBot: The course is graded based on assignments, quizzes, and final project .")

        elif "lecturer" in user_input:
            print("ChatBot: You can reach Lecturers anytime during the week, either through a physical meeting or an email.")

        elif "bye" in user_input:
            print("ChatBot: Goodbye! Good luck with your studies. ")
            break

        else:
            print("ChatBot: Sorry, I didn’t understand that. Can you ask differently?")


course_assistant_bot()
```
